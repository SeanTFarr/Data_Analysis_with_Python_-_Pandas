## Joining DataFrames

#### Inner join

```python
transactions = pd.read_csv("../retail/transactions.csv", parse_dates=["date"])

oil = pd.read_csv("../retail/oil.csv", parse_dates=["date"])
```

```python
transactions.info()
```

```python
oil.info()
```

```python
transactions.merge(oil,
                  how="inner",
                  left_on=["date"],
                  right_on=["date"]
                  )
```

#### Left Join

```python
transactions.merge(oil,
                  how="left",
                  left_on=["date"],
                  right_on=["date"]
                  )#.iloc[trasanctions["date"]== '2013-01-06']
```
