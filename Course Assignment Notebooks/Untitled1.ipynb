```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
```

```python
#lists
names =["Adam born", "Adam died", "Seth born", "Seth died",
       "Enosh born", "Enosh died", "Kenan born", "Kenan died",
       "Mahalalel born", "Mahalalel died", "Jared born", "Jared died",
       "Enoch born", "Enoch died", "Methuselah born", "Methuselah died",
       "Lamech born", "Lamech died", "Noah born", "Noah died",
       "Shem born", "Shem died", "Arpachshad born", "Arpachshad died",
       "Shelah born", "Shelah died", "Eber born", "Eber died",
       "Peleg born", "Peleg died", "Reu born", "Reu died",
       "Serug born", "Serug died", "Nahor born", "Nahor died",
       "Terah born", "Terah died", "Abraham born", "Abraham died",
       "Isaac born", "Isaac died", "Jacob born", "Jacob died"]

years = ["0", "930", "130", "1042",
        "235", "1140", "325", "1235",
        "395", "1290", "460", "1422",
        "622", "987", "687", "1656",
        "874", "1651", "1056", "2006",
        "1556", "2158", "1658", "2096",
        "1693", "2126", "1723", "2187",
        "1757", "1996", "1787", "2026",
        "1819", "2049", "1849", "1997",
        "1878", "2083", "1948", "2123",
        "2048", "2228", "2108", "2255"]
years = [datetime.strptime(d,"%Y") for d in years]
```

```python
# # Choose some nice levels
# levels = np.tile([-5, 5],
#                  int(len(years)/2))[:len(years)]
```

```python
# Choose some nice levels
levels = np.tile([-5, 5],
                 int(len(years)/2))[:len(years)]

# Create figure and plot a stem plot with the date
fig, ax = plt.subplots(figsize=(8.8, 4), layout="constrained")
ax.set(title="Genesis Record of Man")
ax.vlines(years, 0, levels, color="tab:red")  # The vertical stems.
ax.plot(years, np.zeros_like(years), "-o",
        color="k", markerfacecolor="w")  # Baseline and markers on it.

# annotate lines
for d, l, r in zip(years, levels, names):
    ax.annotate(r, xy=(d, l),
                xytext=(-3, np.sign(l)*3), textcoords="offset points",
                horizontalalignment="right",
                verticalalignment="bottom" if l > 0 else "top")
    
# remove y-axis and spines
ax.yaxis.set_visible(False)
ax.spines[["left", "top", "right"]].set_visible(False)


ax.margins(y=0.1)
plt.show()
```
