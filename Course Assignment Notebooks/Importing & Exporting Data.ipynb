```python

url = "https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_population"

pop_data =  pd.read_html(url)[0]
```

```python
pop_data.head()
```

```python
# To drop second "%" column
pop_data=pop_data.drop(pop_data.columns[9], axis=1)

pop_data.head(5)
```

```python
#drop first row and index level
pop_data = (pop_data
           .droplevel(level=0, axis=1)
           #.drop(index=0)
           )
```

```python
pop_data.head(6)
```

```python
pop_data.dtypes
```

```python

pop_data["%"] = pop_data["%"].replace('%','', regex=True)
#.astype(str).astype(float)
```

```python
pop_data.head(6)
```

```python
pop_data["%"] = (pop_data["%"]).astype(float)

```

```python
pop_data.info()
```

```python
pop_data["%_change"] = pop_data["%"].fillna(0).astype(int)
```

```python
pop_data.head(6)
```

```python
pop_data.dtypes
```

```python
pop_data.dtypes
```

```python
# plot the top 10, based on 2022 est., (column 3)
(pop_data
.set_index("State or territory")
.iloc[:10, 3]
.plot.bar());
```
